```python
#importação de biblioteca
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```python
#lendo uma imagem
imagem = cv2.imread('stephenh.jpg',1)

#descobrindo a largura da imagem
print(imagem.shape[1])

#descobrindo a altura da imagem
print(imagem.shape[0])
```

```python
# Plotagem da imagem
plt.imshow(imagem)
```

```python
# gerando histograma geral combinando duas funções: plt.hist() e ravel()
plt.hist(imagem.ravel(), 256, [0,256])
plt.show
```

```python
# A função split da biblioteca cv2 separa os dados dos canais nas variáveis
# azul, verde e vermelho

imagem2 = cv2.imread('stephenh.jpg')

azul, verde, vermelho = cv2.split(imagem2)
```

```python
# Gerando histogramas específicos para cada canal de cor

fig = plt.figure(figsize=(20,5)) # especifica que as imagens plotadas terão o tamanho de 20x5

ax1 = fig.add_subplot(131) # 131 significa 1 linha, três colunas e referente a primeira imagem
ax1.hist(azul.ravel(), 256, [0,256])
plt.title('Histograma da cor Azul')

ax2 = fig.add_subplot(132)
ax2.hist(verde.ravel(), 256, [0,256])
plt.title('Histograma da cor Verde')

ax3 = fig.add_subplot(133)
ax3.hist(vermelho.ravel(), 256, [0,256])
plt.title('Histograma da cor Vermelho')

plt.show()
```

```python
# Equalização do histograma

imagem3 = cv2.imread('stephenh.jpg', 0)

# Equalização do histograma - criamos a variável imagem_eq_hist que por sua vez chama a função 
# cv2.equalizeHist parmetrizada com a imagem 3

imagem_eq_hist = cv2.equalizeHist(imagem3)
```

```python
# Carregando e convertendo as imagens em preto e branco e escalonados em escalas de cinza
# evitando erros de processamento

fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(121)
plt.imshow(imagem3, cmap=plt.cm.gray)

ax2 = fig.add_subplot(122)
plt.imshow(imagem_eq_hist, cmap=plt.cm.gray)

plt.show()

# O resultado será a imagem da esquerda em sua forma netural de preto e branco 
# e à direita, a imgem em preto e branco gerada pela função equalizadora
```

```python
# Gerando os histogramas usando as funções plt.hist() e ravel()

fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(121)
plt.hist(imagem3.ravel(), 256, [0,256])

ax2 = fig.add_subplot(122)
plt.hist(imagem_eq_hist.ravel(), 256, [0,256])

plt.show
```
